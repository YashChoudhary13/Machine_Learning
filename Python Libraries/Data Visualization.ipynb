## Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11) #np.linspace is used to create np array => np.linspace(start,stop,intervals) intervals let's you specify the count of elements to generate
y = x ** 2
print(f"{x} and {y}")
```

```python
plt.plot(x,y) # Creates a line graph
plt.title("Title")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
plt.subplot(2,2,1) # subplot(number of rows, number of columns, plot number) 
plt.plot(x,y)
plt.subplot(2,2,2) # It divides a space into cells/plots depending on arguments 
plt.plot(y,x)
plt.subplot(2,2,3) # and then you can select those plot and make charts in them to display it together
plt.plot(x**2, y**2)
plt.subplot(2,2,4)
plt.plot(x ** 3, y ** 3)
```

**Object Oriented Way**

```python
fig = plt.figure() # Selected empty canvas
# 2 arguments can be passed inside figure(), one figsize=(width, height) which decides the width x height of the figure in inches, and second is dpi = dots per inch (pixel per inch) => higher dpi means higher clarity
axis1 = fig.add_axes([0.1,0.1,0.8,0.8]) #add_axes(, width %, height %)
axis1.plot(x,y)
axis2 = fig.add_axes([0.4,0.4,0.6,0.6])
axis2.plot(x,y)
# First two arguments moves your plot x % from left and y% from bottom
```

## Seaborn

- Quantative variable which cannot be categorized and can have any random numerical value
- Qualitative variable are variables which can be categorized

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Distribution plot

We use distribution plots to visualize the distribution of quantitative data.  
Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:
- distplot/histplot
- jointplot
- pairplot
- rugplot

```python
# Histplot
import pandas as pd
tips = pd.read_csv("tips.csv")
print(tips['size'].unique()) # Hence size is also Qualitative
plt.subplot(1,2,1)
sns.histplot(tips["total_bill"], bins = 50) #By default bins=20 # Type of graph best to show distribution. In each bin, the number of count shows how much values lies in it
plt.subplot(1,2,2)
sns.histplot(tips["tip"])
```

```python
# Jointplot - for checking plot of 2 quantative variables
sns.jointplot(x = 'total_bill', y = 'tip', data = tips) # By default kind = "scatter"
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
# Other kinds = hex, reg
```

```python
# pairplot
sns.pairplot(tips, hue='size', palette='rainbow') # hue lets us change color depending on category
```

```python
# Rug plot
sns.rugplot(tips["total_bill"])
```

### Categorical plot

Now let's discuss using seaborn to plot categorical data!  There are a few main plot types for this:  
- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot

```python
#countplot - which category has more, or which has a particular attribute more
df = pd.read_csv("tips.csv")
sns.countplot(x =df["sex"], hue = df['smoker']) # hue lets us select the category
# Divided between male and female then among them divided into smoker and non smoker
```

```python
# barplot
# sns.barplot(x = df["sex"])
# y-axis -> quantative => average, x-axis -> categorical
import numpy as np
sns.barplot(x = df["sex"], y = df["total_bill"], estimator=np.sum)
```

```python
#box plot
sns.boxplot(x = 'tip', y = 'day', data = df)
```

```python
sns.violinplot(x = 'tip', y = 'day', data = df) # Shows spread, deviation
```

#### stripplot and swarmplot
The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.  
  
The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they dont overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
sns.stripplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.violinplot(x= 'tip',y = 'day',data = df)
sns.swarmplot(x= 'tip',y = 'day',data = df)
```
